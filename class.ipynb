```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 讀取 CSV 檔案，請將 'your_file.csv' 替換成你的檔案名稱
file_path = "./data/bert_labeled_all_other_reviews.csv"
all_df = pd.read_csv(file_path)

all_df
```

```python
#取多分類
all_df['Published Date'] = pd.to_datetime(all_df['Published Date'])

label_cols = ['Label_1', 'Label_2', 'Label_3', 'Label_4', 'Label_5', 'Label_6', 'Label_7', 'Label_8']
threshold = 0.3

# 直接覆蓋原本的 Label 欄位（注意：此操作不可回復，建議先存原始版本）
all_df[label_cols] = (all_df[label_cols] > threshold).astype(int)

all_df
```

```python
df = all_df
# 取得 Label_1 到 Label_8 欄位
label_cols = [f"Label_{i}" for i in range(1, 9)]

# 計算每列中有多少個欄位為 1
df["label_sum"] = df[label_cols].sum(axis=1)

# 篩選出 sum >= 2 的列（代表至少有兩個標記為 1）
multi_label_df = df[df["label_sum"] == 1]

# 顯示前幾筆結果
print(multi_label_df[["Sentence"] + label_cols + ["label_sum"]].head())

# 印出總數量
print("總共有筆數：", len(multi_label_df))


#3筆是三標籤，873筆是二標籤，24239筆是一標籤
```

```python
# 假設 all_df 是你目前的 DataFrame，Label_8 需要移除
label_cols = ['Label_1', 'Label_2', 'Label_3', 'Label_4', 'Label_5', 'Label_6', 'Label_7','Label_8']

# 建立一個新的 DataFrame，用於儲存展開的結果
expanded_rows = []

# 遍歷每一列
for _, row in all_df.iterrows():
    for label in label_cols:
        if row[label] == 1:
            # 複製該列資料，並只保留對應標籤資訊
            new_row = row.drop(label_cols).copy()
            new_row['Label'] = label  # 新增一欄標註該列是屬於哪一個分類
            expanded_rows.append(new_row)

# 建立新的 DataFrame
final_df = pd.DataFrame(expanded_rows)

```

```python

final_df.to_csv("temp.csv")
final_df
# 25994 = 25115+873+3+3 
```
