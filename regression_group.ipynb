```python
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

```python
# 讀取資料
df = pd.read_csv('./0909/final_5.csv')
df
```

```python
cols_to_standardize = [
    'Count', 'Positive Count', 'Negative Count',
    'Count[-1:+1]', 'Positive Count[-1:+1]', 'Negative Count[-1:+1]',
    
    'SAAR_t0','SCAAR[-1,+1]','SCAAR[-1,0]','SCAAR[0,+1]',
    'SAAR_t1',
    'SAAR_t-1',

    'EPS'
]

#測[0,1]
scaler = StandardScaler()
df[cols_to_standardize] = scaler.fit_transform(df[cols_to_standardize])

df


```

```python
# 欄位：自變數 X 與控制變數
x_columns = [
    'Count', 'Avg Sentiment',           #### v_1
     
      'Avg Positive Sentiment', 'Avg Negative Sentiment',
    'Positive Count', 'Negative Count',      #### v_2
    
    'Count[-1:+1]', 'Avg Sentiment[-1:+1]',     #### v_3
     
    'Avg Positive Sentiment[-1:+1]','Avg Negative Sentiment[-1:+1]',
    'Positive Count[-1:+1]', 'Negative Count[-1:+1]', #### v_4

    'Corporation size', 'LEV', 'Current ratio', 'EPS'
]
```

## SAAR_t0 <-> v_1

```python
x_columns =[
      'Avg Positive Sentiment', 'Avg Negative Sentiment',
    'Positive Count', 'Negative Count', 'EPS'     #### model_1
]
# 儲存模型結果
results = {}
```

```python
x_columns =[
    'Avg Positive Sentiment[-1:+1]','Avg Negative Sentiment[-1:+1]',
    'Positive Count[-1:+1]', 'Negative Count[-1:+1]',   #### model_2
    'EPS'
]
# 儲存模型結果
results = {}
```

```python
import statsmodels.api as sm

labels = [f'Label_{i}' for i in range(1, 8)]
results = {}

for label_name in labels:
    df_label = df[df['Label'] == label_name].dropna(subset=['SAAR_t0'] + x_columns)

    if len(df_label) < 10:
        print(f"{label_name} 資料筆數不足，跳過。")
        continue

    X = df_label[x_columns]
    y = df_label['SAAR_t0']
    X = sm.add_constant(X)

    model = sm.OLS(y, X).fit()
    results[label_name] = model

    print(f"\n=== {label_name} 回歸結果（n={len(df_label)}） ===")
    print(f"{'變數':<30} {'Coef':>10} {'P-Value':>10} {'顯著':>6}")
    print('-' * 60)

    for var, coef, pval in zip(model.params.index, model.params.values, model.pvalues.values):
        if pval < 0.01:
            stars = '***'
        elif pval < 0.05:
            stars = '**'
        elif pval < 0.1:
            stars = '*'
        else:
            stars = ''
        print(f"{var:<30} {coef:>10.4f} {pval:>10.4f} {stars:>6}")

    print('-' * 60)
    print(f"{'R-squared':<30} {model.rsquared:>10.4f}")
    print(f"{'Adj. R-squared':<30} {model.rsquared_adj:>10.4f}")
```
